### (5)鋪設繪圖的底圖網格

```python
#產生橫軸、縱軸座標的數據點矩陣(在-5和5之間，間隔0.1)
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
#利用橫軸、縱軸座標的數據點鋪設網點
x,y=np.meshgrid(x,y)
```
