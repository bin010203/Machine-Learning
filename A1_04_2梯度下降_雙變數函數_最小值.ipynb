# 實作04：利用梯度下降法逼近雙變數函數的最小值

### (1)import必要的套件

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as plt_font
twfont1=plt_font.FontProperties(fname="/content/drive/MyDrive/Colab Notebooks/kaiu.ttf")
```

### (2)定義雙變數函數f(x,y)=x<sup>2</sup>+y<sup>2</sup>學生可以自行修改)

```python
def F(x,y):
    return x**2+y**2
```

### (3)定義雙變數函數的偏導函數fx=2x及fx=2y(學生可以自行修改)

```python
def Fx(x,y):
    return 2*x
def Fy(x,y):
    return 2*y
```

### (4)利用梯度下降找函數的極小值

```python
#設定step_x、step_y用來儲存梯度下降每一步的座標
step_x=[]
step_y=[]
step_z=[]
#設定學習率，可以調大或調小，觀察梯度下降情況
ETA=0.1
#設定梯度下降的起點，可以改變不同位置，觀察梯度下降情況
x_point=-3
y_point=-5
counter = 0
while abs(Fx(x_point,y_point)) >= 0.0001 or abs(Fy(x_point,y_point)) >= 0.0001:
    step_x.append(x_point)
    step_y.append(y_point)
    step_z.append(F(x_point,y_point))
    #利用梯度下降，更新x_point和y_point
    x_point=x_point-ETA*Fx(x_point,y_point)
    y_point=y_point-ETA*Fy(x_point,y_point)
    counter += 1
```

```python
 print("下降次數 = ",counter)
```

### (5)鋪設繪圖的底圖網格

```python
#產生橫軸、縱軸座標的數據點矩陣(在-5和5之間，間隔0.1)
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
#利用橫軸、縱軸座標的數據點鋪設網點
x,y=np.meshgrid(x,y)
```

### (6)劃出曲面圖及梯度下降路徑

```python
#開啟繪圖區域
fig=plt.figure(figsize=(12,6))
#設定為3D繪圖
ax=fig.gca(projection='3d')
#設定坐標軸範圍
ax.set_zlim(0,50)
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
#設定坐標軸說明文字
ax.set_xlabel('X軸',fontproperties=twfont1)
ax.set_ylabel('Y軸',fontproperties=twfont1)
ax.set_zlabel('Z軸',fontproperties=twfont1)
#畫出3D曲面,cmap為數值及色階對應方式,linewidth為線寬,antiliased=Fasle可以加快繪圖速度 
surf=ax.plot_surface(x,y,F(x,y),cmap="hot",alpha=0.2,antialiased=False)
# 畫出梯度修正路徑的點和線
ax.plot(step_x,step_y,step_z,"k:o");
# 顯示數值及色階對應方式,shrink設定colorbar長度為圖片高度的幾倍,預設值1.0,aspect設定colorbar寬、高比,預設值20
fig.colorbar(surf,shrink=0.5,aspect=5)
#elev=20設定視線和水平面的夾20度，azim=30設定立體圖順時針旋轉30度
ax.view_init(elev=50,azim=10)
```
