### (4)劃出曲面圖

```python
#開啟繪圖區域
fig=plt.figure(figsize=(12,9))
#設定為 3D 繪圖
ax=fig.gca(projection='3d')
#設定z軸範圍
ax.set_zlim(0,40)
#設定坐標軸名稱
ax.set_xlabel('X軸',fontproperties=twfont1)
ax.set_ylabel('Y軸',fontproperties=twfont1)
ax.set_zlabel('Z軸',fontproperties=twfont1)
#畫出3D曲面,cmap為數值及色階對應方式,antialiased=Fasle可以加快繪圖速度 
surf=ax.plot_surface(x,y,F(x,y),cmap="cool",alpha=0.5,antialiased=False)
#顯示數值及色階對應方式,shrink設定colorbar長度為圖片高度的幾倍,預設值1.0,aspect設定colorbar寬、高比,預設值20
fig.colorbar(surf,shrink=0.5,aspect=5)
#elev=20設定視線和水平面的夾20度，azim=30設定立體圖順時針旋轉30度
ax.view_init(elev=40,azim=45)
```

```python
#開啟繪圖區域
fig=plt.figure(figsize=(12,9))
#設定為 3D 繪圖
ax=fig.gca(projection='3d')
#設定z軸範圍
ax.set_zlim(0,40)
#設定坐標軸名稱
ax.set_xlabel('X軸',fontproperties=twfont1)
ax.set_ylabel('Y軸',fontproperties=twfont1)
ax.set_zlabel('Z軸',fontproperties=twfont1)
#畫出3D曲面,cmap為數值及色階對應方式,antialiased=Fasle可以加快繪圖速度 
surf=ax.plot_surface(x,y,F(x,y),cmap="hsv",alpha=0.5,antialiased=False)
#顯示數值及色階對應方式,shrink設定colorbar長度為圖片高度的幾倍,預設值1.0,aspect設定colorbar寬、高比,預設值20
fig.colorbar(surf,shrink=0.5,aspect=5)
#elev=20設定視線和水平面的夾20度，azim=30設定立體圖順時針旋轉30度
ax.view_init(elev=60,azim=45)
```

### (5)劃出等高線

```python
# 開啟繪圖區域
plt.figure(figsize=(5,5))
plt.title('等高線圖',fontproperties=twfont1,fontsize=20)
CS=plt.contour(x,y,F(x,y),50)
plt.clabel(CS, inline=1, fontsize=10)
```
