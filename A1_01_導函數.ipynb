$L=\lim _{h \rightarrow 0} \frac{f(a+h)-f(a)}{h}$

$$L=\lim _{h \rightarrow 0} \frac{f(a+h)-f(a)}{h}$$

Solution: Since the derivative of the parabola $f^{\prime}(x)=2 a_{2} x+a_{1}$ then $f^{\prime}(c)=2 a_{2} c+a_{1}$, and
$$
\begin{array}{c}
\frac{f(b)-f(a)}{b-a}=\frac{a_{2} b^{2}+a_{1} b+a_{0}-\left(a_{2} a^{2}+a_{1} a+a_{0}\right)}{b-a}=\frac{(b-a)\left[a_{2}(a+b)+a_{1}\right]}{b-a}=a_{2}(a+b)+a_{1}, \\
f^{\prime}(c)=\frac{f(b)-f(a)}{b-a}=a_{2}(a+b)+a_{1}, \quad f^{\prime}(c)=2 a_{2} c+a_{1}=a_{2}(a+b)+a_{1}, \quad c=\frac{a+b}{2}
\end{array}
$$
